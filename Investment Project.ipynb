```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
```

```python
data = {
    'Investment_Type': ['Stocks', 'Bonds', 'Real_Estate', 'Mutual_Funds', 'Cryptocurrency', 'Gold', 'Savings_Account'],
    'Initial_Investment': [10000, 5000, 20000, 8000, 3000, 15000, 1000],
    'Investment_Duration_Years': [5, 10, 7, 3, 2, 8, 1],
    'Expected_Return_Percentage': [8, 4, 6, 7, 15, 5, 1],
    'Reason_for_Investment': ['Growth', 'Security', 'Growth', 'Growth', 'Speculation', 'Security', 'Liquidity']
}
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Investment_Type', y='Initial_Investment', data=df, palette='viridis')
plt.title('Initial Investment by Investment Type')
plt.xlabel('Investment Type')
plt.ylabel('Initial Investment ($)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Investment_Type', y='Expected_Return_Percentage', data=df, palette='coolwarm')
plt.title('Expected Return Percentage by Investment Type')
plt.xlabel('Investment Type')
plt.ylabel('Expected Return (%)')
plt.xticks(rotation=45)
plt.show()
```

```python
correlation_matrix = df[['Initial_Investment', 'Investment_Duration_Years', 'Expected_Return_Percentage']].corr()
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Initial_Investment', y='Expected_Return_Percentage', hue='Investment_Type', data=df, palette='viridis', s=100)
plt.title('Initial Investment vs Expected Return Percentage')
plt.xlabel('Initial Investment ($)')
plt.ylabel('Expected Return (%)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Investment_Duration_Years', y='Expected_Return_Percentage', hue='Investment_Type', data=df, palette='viridis', s=100)
plt.title('Investment Duration vs Expected Return Percentage')
plt.xlabel('Investment Duration (Years)')
plt.ylabel('Expected Return (%)')
plt.show()
```

```python
corr, _ = pearsonr(df['Initial_Investment'], df['Expected_Return_Percentage'])
print(f'Pearson correlation between Initial Investment and Expected Return Percentage: {corr:.2f}')
```

```python
corr, _ = pearsonr(df['Investment_Duration_Years'], df['Expected_Return_Percentage'])
print(f'Pearson correlation between Investment Duration and Expected Return Percentage: {corr:.2f}')
```
